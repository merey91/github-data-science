<h1>Opening a New Chinese Restaurant in Toronto,Canada | Introduction and Business Problems<h1>

<h2>Introduction:<h2>

This is a capstone project for IBM Data Science Professional Certificate. In this project, I am creating a hyphothetical scenario for a concept that there may not enough Chinese Restaurant in Toronto,Canada.Therefore it might be  a great opportunity for an entrepreneur who is based in Canada. As the Chinese Food is popular among Asian community, so this entrepreneur might think of opening its business in areas where asian community resides. With the purpose in mind, finding the location to open such a restaurant is one of the most important decisions for this entrepreneur and I am designing this project to help him find the most suitable location. 

<h2>Business Problem:<h2>

The objective of this capstone project is to find the most suitable location for the entrepeneur to open a new Chinese restaurant in Toronto, Canada. By using data science methods and tools along with mechine learning algorithms such as clustering,  this project aims to provide solutions to answer the business question: In Toronto, if  entrepeneur is looking to open a Chinese restaurant, where should they consider opening it? 

To find the answers to the following questions: 

<ol>
<li>List and Visualize all major parts of Toronto that has great Chinese restaurants. </li>
<li>What is best location in Toronto for Chinese cuisine? </li>
<li>Which areas have potential Chinese restaurant Market? </li>
<li>Which all areas lack Chinese Restaurants? </li>
<li>Which is the best place to stay if you prefer Chinese cuisine? </li>  
</ol>

<h2>Target Audience:<h2>

The  entrepeneur who wants to find the location to open authentic Chinese restaurant.
